# deepface
Deepface is a lightweight face recognition and facial attribute analysis (age, gender, emotion and race) framework for python. It is a hybrid face recognition framework wrapping state-of-the-art models: VGG-Face, Google FaceNet, OpenFace, Facebook DeepFace, DeepID, ArcFace and Dlib. The library is mainly based on Keras and TensorFlow.

### import Package

```python
import matplotlib.pyplot as plt
from deepface import DeepFace
import time
```

### Face Verification
Verification function under the deepface interface offers to verify face pairs as same person or different persons

```python
models = ["VGG-Face", "Facenet", "OpenFace", "DeepFace", "DeepID", "ArcFace", "Dlib"]
result  = DeepFace.verify("images/1.jpg", "images/2.jpg", model_name = models[0])
result
```

```python
models = ["VGG-Face", "Facenet", "OpenFace", "DeepFace", "DeepID", "ArcFace", "Dlib"]
result  = DeepFace.verify("images/1.jpg", "images/7.jpg", model_name = models[0])
result
```




### Facial Attribute Analysis
Deepface also offers facial attribute analysis including age, gender, facial expression (including angry, fear, neutral, sad, disgust, happy and surprise) and race (including asian, white, middle eastern, indian, latino and black) predictions. 

```python
obj = DeepFace.analyze(img_path = "images/7.jpg", actions = ['age', 'emotion'])
```

```python
obj
```

### Face Detectors
Face detection and alignment are early stages of a modern face recognition pipeline. OpenCV, SSD, Dlib, MTCNN and RetinaFace methods are wrapped in deepface as a facial detector. 

```python
backends = ['opencv', 'ssd', 'dlib', 'mtcnn', 'retinaface']
detected_face = DeepFace.detectFace("images/7.jpg", detector_backend = backends[0])
```

```python
plt.imshow(detected_face)
```
